```python
inert=[]
for i in range(2,11):
    model_i=KMeans(n_clusters=i).fit(dummy)
    inert.append(model_i.inertia_)
```

```python
plt.figure(figsize=(18,6))
plt.plot(range(2,11),inert,marker='o')
plt.grid()
plt.xlabel('N_clusters')
plt.ylabel('Inertia')
```

```python
from sklearn.decomposition import PCA
```
